# Importación de datos

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 1. Análisis de facturación

Ingresos totales por tienda

```python
ingreso_t1 = tienda['Precio'].sum()
ingreso_t2 = tienda2['Precio'].sum()
ingreso_t3 = tienda3['Precio'].sum()
ingreso_t4 = tienda4['Precio'].sum()
print(f"Ingresos Tienda 1: ${ingreso_t1:,.2f}")
print(f"Ingresos Tienda 2: ${ingreso_t2:,.2f}")
print(f"Ingresos Tienda 3: ${ingreso_t3:,.2f}")
print(f"Ingresos Tienda 4: ${ingreso_t4:,.2f}")
```

```python
resumen_ingresos = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_t1, ingreso_t2, ingreso_t3, ingreso_t4]
})

print("\n--- Tabla de Resumen ---")
print(resumen_ingresos)
```

# 2. Cantidad de productos vendidos por categoría

```python
ventas_t1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_t2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_t3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_t4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("--- Categorías más populares Tienda 1 ---")
print(ventas_t1.head())

print("\n--- Categorías más populares Tienda 2 ---")
print(ventas_t2.head())

print("\n--- Categorías más populares Tienda 3 ---")
print(ventas_t3.head())

print("\n--- Categorías más populares Tienda 4 ---")
print(ventas_t4.head())
```

# 3. Calificaciones promedios de los clientes por tienda

```python
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()

print(f"Satisfacción promedio Tienda 1: {promedio_t1:.2f} ⭐")
print(f"Satisfacción promedio Tienda 2: {promedio_t2:.2f} ⭐")
print(f"Satisfacción promedio Tienda 3: {promedio_t3:.2f} ⭐")
print(f"Satisfacción promedio Tienda 4: {promedio_t4:.2f} ⭐")

resumen_satisfaccion = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [promedio_t1, promedio_t2, promedio_t3, promedio_t4]
}).sort_values(by='Calificación Promedio', ascending=False)

print("\n--- Ranking de Satisfacción ---")
print(resumen_satisfaccion)
```

# 4. Productos más y menos vendidos



```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, df in enumerate(tiendas):
    conteo_productos = df['Producto'].value_counts()

    mas_vendidos = conteo_productos.head(5)
    menos_vendidos = conteo_productos.tail(5)

    print(f"\n--- {nombres_tiendas[i]} ---")
    print("Top 5 Más Vendidos:")
    print(mas_vendidos)
    print("\nTop 5 Menos Vendidos:")
    print(menos_vendidos)

    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    mas_vendidos.plot(kind='bar', color='skyblue')
    plt.title(f'Más Vendidos - {nombres_tiendas[i]}')
    plt.ylabel('Cantidad')

    plt.subplot(1, 2, 2)
    menos_vendidos.plot(kind='bar', color='salmon')
    plt.title(f'Menos Vendidos - {nombres_tiendas[i]}')

    plt.tight_layout()
    plt.show()
```

# 5. Costo de envio promedio por tienda

```python
costo_envio_t1 = tienda['Costo de envío'].mean()
costo_envio_t2 = tienda2['Costo de envío'].mean()
costo_envio_t3 = tienda3['Costo de envío'].mean()
costo_envio_t4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio Tienda 1: ${costo_envio_t1:.2f}")
print(f"Costo de envío promedio Tienda 2: ${costo_envio_t2:.2f}")
print(f"Costo de envío promedio Tienda 3: ${costo_envio_t3:.2f}")
print(f"Costo de envío promedio Tienda 4: ${costo_envio_t4:.2f}")

resumen_envios = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Envío Promedio': [costo_envio_t1, costo_envio_t2, costo_envio_t3, costo_envio_t4]
})

print("\n--- Resumen Comparativo de Envíos ---")
print(resumen_envios)
```

# Visualización

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (15, 5)

# 1.  Ingresos Totales por Tienda
plt.subplot(1, 3, 1) # 1 fila, 3 columnas, posición 1
sns.barplot(x='Tienda', y='Ingreso Total', data=resumen_ingresos, palette='viridis')
plt.title('Ingresos Totales por Tienda', fontsize=14)
plt.ylabel('Dólares ($)')
plt.xlabel('')

# 2. Satisfacción del Cliente (Tendencia de Calidad)
plt.subplot(1, 3, 2) # Posición 2

resumen_satisfaccion_sorted = resumen_satisfaccion.sort_values('Tienda')
plt.plot(resumen_satisfaccion_sorted['Tienda'], resumen_satisfaccion_sorted['Calificación Promedio'],
         marker='o', linestyle='-', color='orange', linewidth=2)
plt.fill_between(resumen_satisfaccion_sorted['Tienda'], resumen_satisfaccion_sorted['Calificación Promedio'], alpha=0.2, color='orange')
plt.title('Nivel de Satisfacción (Promedio)', fontsize=14)
plt.ylim(0, 5) # Las estrellas suelen ser de 0 a 5
plt.ylabel('Puntuación')

# 3. Relación Precio vs Costo de Envío
# Usaremos los datos de la Tienda 1 como muestra representativa
plt.subplot(1, 3, 3) # Posición 3
sns.scatterplot(x='Precio', y='Costo de envío', data=tienda, alpha=0.6, color='purple')
plt.title('Relación Precio vs Envío (Tienda 1)', fontsize=14)
plt.xlabel('Precio del Producto')
plt.ylabel('Costo de Envío')

# Ajustar diseño para que no se amontonen
plt.tight_layout()
plt.show()
```

# Visualización Geografica

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(15, 12))
tiendas_list = [tienda, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['blue', 'green', 'orange', 'red']

for i, df in enumerate(tiendas_list):
    ax = axes[i//2, i%2]
    ax.scatter(df['lon'], df['lat'], alpha=0.5, c=colores[i], s=10)
    ax.set_title(f'Distribución Geográfica - {nombres[i]}')
    ax.set_xlabel('Longitud')
    ax.set_ylabel('Latitud')

plt.tight_layout()
plt.show()
```

# **Análisis Geográfico y Patrones de Venta**

Tras mapear las coordenadas de latitud (lat) y longitud (lon), se identificaron los siguientes patrones:

Densidad Comercial: Las tiendas con mayores ingresos (Tienda 1 y 2) presentan clústeres de ventas más definidos. Esto indica que operan en zonas con alta demanda concentrada, lo que facilita el posicionamiento de marca.

Debilidad de la Tienda 4: La distribución geográfica de la Tienda 4 muestra una dispersión mayor. A pesar de tener el costo de envío más bajo ($23,459.46), no logra capturar mercados de alta densidad, lo que limita su crecimiento y explica sus bajos ingresos totales.

Correlación Geográfica-Económica: Se observa que las áreas con mayor concentración de ventas también presentan las mejores calificaciones de satisfacción, sugiriendo que la cercanía o la densidad de operaciones mejora la experiencia del cliente.

Conclusión Geográfica: La ubicación de las ventas de la Tienda 4 no es estratégica para el volumen de negocio que el Sr. Juan requiere, reafirmando la decisión de venta de este activo.
